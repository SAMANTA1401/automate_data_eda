```python
import pytesseract
from PIL import Image
import io
```

```python
def process_document(file_bytes: bytes) -> str:
    image = Image.open(io.BytesIO(file_bytes))
    text = pytesseract.image_to_string(image)
    return text
```

```python
def extract_text_with_tables(pdf_path):
    all_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            all_text += page.extract_text()
    return all_text
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(text: str):
    doc = nlp(text)
    extracted = {}
    for ent in doc.ents:
        extracted[ent.label_] = ent.text
    return extracted
```

```python
from jinja2 import Template
import pdfkit

def generate_report(data: dict) -> str:
    with open("templates/report_template.html") as f:
        html = f.read()
    report = Template(html).render(data=data)
    pdfkit.from_string(report, "output/report.pdf")
    return "output/report.pdf"
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication

def send_report_via_email(to, subject, pdf_path):
    msg = MIMEMultipart()
    msg['To'] = to
    msg['Subject'] = subject

    with open(pdf_path, 'rb') as f:
        attach = MIMEApplication(f.read(), _subtype="pdf")
        attach.add_header('Content-Disposition','attachment', filename="report.pdf")
        msg.attach(attach)

    server = smtplib.SMTP('smtp.yourdomain.com', 587)
    server.starttls()
    server.login("you@example.com", "yourpassword")
    server.sendmail("you@example.com", to, msg.as_string())
    server.quit()
```
